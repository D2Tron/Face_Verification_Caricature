```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv("./wandb_csv/id_sub_loss.csv")
```

```python
# Filter out columns that end with "F1 Score"
f1_score_columns = [col for col in df.columns if col.endswith('Validation')]
```

```python
# Define custom legend labels for each line
legend_labels = {
    "sub car fold1 - Validation": "Caricature (Fold 1)",
    "sub ver fold3 - Validation": "Veridical (Fold 3)",
    "sub carver fold1 - Validation": "Combined (Fold 1)"
}
```

```python
plt.figure(figsize=(10, 6))  # Adjust figsize as needed

# Plotting each metric
for metric_col in f1_score_columns:
    plt.plot(df['Step'], df[metric_col], label=legend_labels.get(metric_col, metric_col))  

plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Validation Loss')
plt.legend()
plt.grid(True)
plt.show()
```
