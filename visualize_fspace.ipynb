```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load feature vectors from .npy files
features_real = np.load('outputs_ver_prom.npy')
features_caricature = np.load('outputs_car_prom.npy')

# Combine the feature vectors and create labels
all_features = np.vstack((features_real, features_caricature))
labels = np.hstack((np.zeros(features_real.shape[0]), np.ones(features_caricature.shape[0])))

# Apply PCA for dimensionality reduction to 2 dimensions
pca = PCA(n_components=2)
embedded_features = pca.fit_transform(all_features)

# Scatter plot of the reduced dimensions
plt.figure(figsize=(10, 6))
scatter = plt.scatter(embedded_features[:, 0], embedded_features[:, 1], c=labels, cmap='viridis')
plt.title('PCA Visualization of Feature Space')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(*scatter.legend_elements(), title='Classes', labels=['Real', 'Caricature'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Load feature vectors from .npy files
features_real = np.load('fspace_ver.npy')
features_caricature = np.load('fspace_car.npy')

# Combine the feature vectors and create labels
all_features = np.vstack((features_real, features_caricature))
labels = np.hstack((np.zeros(features_real.shape[0]), np.ones(features_caricature.shape[0])))

# Apply t-SNE for dimensionality reduction to 2 dimensions
tsne = TSNE(n_components=2, random_state=42)
embedded_features = tsne.fit_transform(all_features)

# Scatter plot of the reduced dimensions
plt.figure(figsize=(10, 6))
scatter = plt.scatter(embedded_features[:, 0], embedded_features[:, 1], c=labels, cmap='viridis')
plt.title('t-SNE Visualization of Feature Space')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.legend(*scatter.legend_elements(), title='Classes', labels=['Real', 'Caricature'])
plt.show()
```
