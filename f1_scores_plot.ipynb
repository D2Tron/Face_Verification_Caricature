```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Import the CSV file in Jupyter Notebook
# df_train = pd.read_csv('./saved_lists/f1_train_ver_prom.csv')
df_val = pd.read_csv('./saved_lists/f1_lists/f1_carver_sub_id.csv')

# Increase figure size and rotate x-axis labels
plt.figure(figsize=(20, 8))
sns.barplot(x='Feature Name', y='F1 Score', data=df_val)
plt.xticks(rotation=90)

plt.title('Validation F1 Scores for Different Features')
plt.xlabel('Feature Name')
plt.ylabel('F1 Score')
plt.show()
```

```python
# Import the CSV file in Jupyter Notebook
# df_val = pd.read_csv('f1_val_.csv')

# Extracting prominent feature information
df_val['Prominent Feature'] = df_val['Feature Name'].apply(lambda x: x.split('_')[0])

# Sort the DataFrame alphabetically based on Feature Name
df_val = df_val.sort_values(by='Feature Name')

# Get unique prominent features
unique_prominent_features = df_val['Prominent Feature'].unique()

# Create a color palette for subgroups (prominent features)
subgroup_palette = sns.color_palette("Set3", n_colors=len(unique_prominent_features))

# Map prominent feature colors to the DataFrame
prominent_feature_color_mapping = dict(zip(unique_prominent_features, subgroup_palette))
df_val['Color'] = df_val['Prominent Feature'].map(prominent_feature_color_mapping)

# Create a bar plot with different colors for each subgroup
plt.figure(figsize=(20, 8))
sns.barplot(x='Feature Name', y='F1 Score', data=df_val, palette=df_val['Color'])
plt.title('Validation F1 Scores for Different Features')
plt.xlabel('Feature Name')
plt.ylabel('F1 Score')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df_val = pd.read_csv('./saved_lists/f1_lists/acc_car_prom_img.csv')

# Filter out rows where Frequency Count is 0
df_val = df_val[df_val['Frequency Count'] > 0]

# Sort the DataFrame alphabetically based on Feature Name
df_val = df_val.sort_values(by='Frequency Count', ascending=False)

# Define color palette
palette = sns.color_palette("plasma", len(df_val))

# Create a figure with one subplot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot F1 scores on the primary y-axis
sns.barplot(x='Feature Name', y='F1 Score', data=df_val, ax=ax1, alpha=0.6, palette=palette)
ax1.set_title('Accuracy per Attribute sorted by Attribute Frequency')
ax1.set_xlabel('Attribute Name')
ax1.set_ylabel('Accuracy')

# Create a secondary y-axis for feature counts on the right-hand side
ax2 = ax1.twinx()
sns.barplot(x='Feature Name', y='Frequency Count', data=df_val, ax=ax2, alpha=1.0, palette=palette)
ax2.set_ylabel('Frequency Count')
ax2.set_ylim(0, 250)  # Set the right-hand side y-axis limit

# Rotate x-axis labels for better visibility on both subplots
ax1.tick_params(axis='x', rotation=90)
ax2.tick_params(axis='x', rotation=90)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
