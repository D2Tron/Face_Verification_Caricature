```python
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

folder_path = './CaricatureSharedDrive/caricature data/new-cropped/barack_obama_caricature/'

# Get a list of all files in the folder
files = os.listdir(folder_path)

# Filter JPEG files
image_paths = [os.path.join(folder_path, file) for file in files if file.endswith('.jpg') or file.endswith('.jpeg')]

# Create a figure and axes
fig, axes = plt.subplots(1, len(image_paths[:5]), figsize=(15, 3))  # Adjust figsize as needed

# Loop through each image path and plot it on its respective axis
for ax, img_path in zip(axes, image_paths[5:10]):
    img = mpimg.imread(img_path)
    ax.imshow(img)
    ax.axis('off')  # Turn off axes

plt.tight_layout()  # Adjust spacing between subplots if needed
plt.show()
```
